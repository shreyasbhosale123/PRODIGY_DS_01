```python
import pandas as pd

df = pd.read_csv("/content/population.csv.csv", skiprows=4)

df.head()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.hist(df["2022"], bins=30)

plt.title("Distribution of Population in 2022")
plt.xlabel("Population")
plt.ylabel("Number of Countries")

plt.show()
```

```python
top10 = df.sort_values(by="2022", ascending=False).head(10)

plt.figure(figsize=(10,6))

plt.bar(top10["Country Name"], top10["2022"])

plt.title("Top 10 Countries by Population (2022)")
plt.xlabel("Country")
plt.ylabel("Population")

plt.xticks(rotation=45)

plt.show()
```
